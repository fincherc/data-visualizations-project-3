```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('../../AustinTxAccidentsData.csv')
```

```python
# Filter the dataset for sunny and overcast conditions
sunny_conditions = ['Clear', 'Fair', 'Fair / Windy']
overcast_conditions = ['Overcast', 'Cloudy', 'Mostly Cloudy']
```

```python
# Create two subsets of the data
sunny_df = df[df['Weather_Condition'].isin(sunny_conditions)]
overcast_df = df[df['Weather_Condition'].isin(overcast_conditions)]
```

```python
# Count the number of accidents in each condition
sunny_accidents = sunny_df.shape[0]
overcast_accidents = overcast_df.shape[0]
```

```python
# Create a bar chart to compare
labels = ['Sunny', 'Overcast']
accident_counts = [sunny_accidents, overcast_accidents]
```

```python

plt.figure(figsize=(8, 6))
plt.bar(labels, accident_counts, color=['orange', 'gray'])
plt.title('Number of Accidents: Sunny vs Overcast Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.show()
```
